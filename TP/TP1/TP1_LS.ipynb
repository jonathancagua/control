```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

muestra = pd.read_csv('daq_log_ls.csv')

```

```python
SAMPLES = 191
adc_float = muestra[1:SAMPLES]['ADC'].apply(lambda x: float(x))
dac_float = muestra[1:SAMPLES]['DAC'].apply(lambda x: float(x))
```

```python

plt.figure(figsize=(24, 6))
plt.title('STEP response')
plt.plot(adc_float[1:SAMPLES], label='ADC')
plt.plot(dac_float[1:SAMPLES], label='DAC')
plt.ylabel('Magnitude')
plt.xlabel('Sample [No]')
plt.legend()
plt.grid()
```
