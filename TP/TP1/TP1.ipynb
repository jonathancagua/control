```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
SAMPLES = 200

muestra = pd.read_csv('daq_log.csv')
muestra

```

```python

plt.figure(figsize=(24, 6))
plt.title('STEP response')
plt.plot(muestra['No'][:SAMPLES], muestra['ADC'][:SAMPLES], 'o', label='ADC')
plt.plot(muestra['No'][:SAMPLES], muestra['DAC'][:SAMPLES], 'o', label='DAC')
plt.ylabel('Magnitude')
plt.xlabel('Sample [No]')
plt.legend()
plt.grid()
```
